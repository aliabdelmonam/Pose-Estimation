```python
import cv2
import mediapipe as mp
import time 
# Initialize Mediapipe Hand module
mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils

# Set up capture
cap = cv2.VideoCapture(0)  # Use webcam
hands = mp_hands.Hands()

pTime = 0
cTime = 0

while True:
    ret,frame = cap.read()
    if not ret:
        break
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result=hands.process(frame_rgb)
    print(result.multi_hand_landmarks)
    

    if result.multi_hand_landmarks:
        # for hand_landmarks in result.multi_hand_landmarks:
        #     mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        #     hand = result.multi_hand_landmarks[0].classification[0].label
        #     index_finger_tip = hand_landmarks.landmark[8]
        #     index_finger_mcp = hand_landmarks.landmark[5]
        for idx, hand_landmarks in enumerate(result.multi_hand_landmarks):
        # Get Handedness Information
            handedness = result.multi_handedness[idx].classification[0].label  # "Left" or "Right"
            print(f"Detected: {handedness} Hand")
            
            # Draw hand landmarks
            mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

            # Display hand type on screen
            cv2.putText(frame, f"{handedness} Hand", (50, 100 + idx * 30), cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 2)
            # if index_finger_tip.y < index_finger_mcp.y:
            #     print("Index finger is raised")
            #     cv2.putText(frame,"Index finger is raised",(200,50),cv2.FONT_HERSHEY_PLAIN,2,(255,0,0),2,cv2.LINE_AA)

            # else:
            #     print("Index finger is closed")
            #     cv2.putText(frame,"Index finger is closed",(200,50),cv2.FONT_HERSHEY_PLAIN,2,(255,0,0),2,cv2.LINE_AA)
    
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(frame,str(int(fps)),(15,60),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0))
    frame = cv2.resize(frame, (1000, 720))  # Resize to 1280x720
    cv2.imshow("Hand Tracking", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
